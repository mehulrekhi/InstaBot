```python
def lst_li_to_href(lst):
    #This function is used for creating list of href from given list.
    href_lst=[]
    for user in lst:
        userLink = user.find_element_by_css_selector('a').get_attribute('href')
        href_lst.append(userLink)
    return href_lst
```

```python
def open_followerlist():
    #This function is used for opening followers list of search_item IG page.
    open_followers=driver.find_element_by_xpath('//a[(@class="-nal3 ")]')  
    open_followers.click()
```

```python
def insta_User_followerlist(max):
    #This function is used for extracting followers list of search_item IG page.
    open_followerlist()
    
    #wait until dialog box of followers is located
    wait=WebDriverWait(driver,20) 
    followersList= wait.until(EC.presence_of_element_located((By.XPATH,'//div[(@role="dialog")]/div[2]/ul/div')))
    
    #extracting the length of followersList
    numberOfFollowersInList = len(followersList.find_elements_by_css_selector('li'))
    followersList.click()
    
    #for scrolling dialogbox
    actionChain = webdriver.ActionChains(driver)
    j=0
    while (numberOfFollowersInList < max):
        #performing pressing space bar on keyboard(key_down) and unpressing it(key_up)
        actionChain.key_down(Keys.SPACE).key_up(Keys.SPACE).perform()
        time.sleep(3)
        
        #extracting the length of followersList
        numberOfFollowersInList = len(followersList.find_elements_by_css_selector('li'))
        j+=1
        if j==1:
            followersList= wait.until(EC.presence_of_element_located((By.XPATH,'//div[(@role="dialog")]//div[(@class="isgrP")]/ul/div')))
            followersList.click()
            
    #creating list of followers href of length max 
    followers = lst_li_to_href(followersList.find_elements_by_css_selector('li')[:max])
        
    #close dialog box
    #close_dialog_box()
    
    #returning list of followers
    return(followers)

def print_user_name_from_url(lst):
    #This function is used for printing the followers username from given url.
    for i in lst:
        print(i.split('/')[3])  
```

```python
FoodTalk_followers=insta_User_followerlist(500)
print_user_name_from_url(FoodTalk_followers)
```
